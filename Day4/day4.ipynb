```python
import pandas as pd
```

```python
input = pd.read_csv('day4_input.txt',header=None)
input = input.set_axis(['elf1','elf2'],axis='columns')
```

```python
input[['elf1_start','elf1_end']]= input.elf1.str.split('-',expand=True).astype(int)
input[['elf2_start','elf2_end']]= input.elf2.str.split('-',expand=True).astype(int)
```

```python
input['elf1_start_kleiner'] = input.apply(
    lambda row: 'yes' if row.elf1_start <= row.elf2_start
                else 'no',
    axis=1    
)
```

```python
input['elf1_end_größer'] = input.apply(
    lambda row: 'yes' if row.elf1_end >= row.elf2_end
                else 'no',
    axis=1    
)
```

```python
input['elf2_start_kleiner'] = input.apply(
    lambda row: 'yes' if row.elf2_start <= row.elf1_start
                else 'no',
    axis=1    
)
```

```python
input['elf2_end_größer'] = input.apply(
    lambda row: 'yes' if row.elf2_end >= row.elf1_end
                else 'no',
    axis=1    
)
```

```python
input['both'] = input.apply(
    lambda row: 1 if (row.elf1_start_kleiner == 'yes' and row.elf1_end_größer == 'yes') or (row.elf2_start_kleiner == 'yes' and row.elf2_end_größer == 'yes')
                else 0,
    axis=1    
)
```
